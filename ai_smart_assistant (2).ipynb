```python
# from openai import OpenAI

import os
from openai import OpenAI

client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))

while True:
    print("\n--- AI ASSISTENT ---")
    print("1 = Frage beantworten")
    print("2 = Email schreiben")
    print("3 = Text verbessern")
    print("4 = Zusammenfassung")
    print("5 = Social Media Post")
    print("6 = Produktbeschreibung")
    print("exit = Beenden")
    user_input = input("Deine Auswahl: ")

    if user_input == "exit":
        print("Programm beendet.")
        break

    
    # FRAGE STELLEN
    
    elif user_input == "1":
        frage = input("Stelle deine Frage: ")

        print("Sprache wählen:")
        print("1 = Deutsch")
        print("2 = Englisch")
        language = input("Deine Wahl: ")

        if language == "2":
            prompt = f"Answer the following question in English:\n{frage}"
        else:
            prompt = f"Beantworte folgende Frage auf Deutsch:\n{frage}"

    
    #  EMAIL GENERIEREN
    
    elif user_input == "2":
        frage = input("Was soll die Email enthalten? ")

        print("Sprache wählen:")
        print("1 = Deutsch")
        print("2 = Englisch")
        language = input("Deine Wahl: ")

        if language == "2":
            prompt = f"Write a professional email in English:\n{frage}"
        else:
            prompt = f"Schreibe eine professionelle Email auf Deutsch:\n{frage}"

    
    #  TEXT VERBESSERN
    
    elif user_input == "3":
        frage = input("Text eingeben: ")

        print("Sprache wählen:")
        print("1 = Deutsch")
        print("2 = Englisch")
        language = input("Deine Wahl: ")

        if language == "2":
            prompt = f"Improve the following text in English:\n{frage}"
        else:
            prompt = f"Verbessere diesen Text auf Deutsch:\n{frage}"
    elif user_input == "4":
        frage = input("Text zum Zusammenfassen eingeben: ")

        print("Sprache wählen:")
        print("1 = Deutsch")
        print("2 = Englisch")
        language = input("Deine Wahl: ")

        if language == "2":
            prompt = f"Summarize the following text in English:\n{frage}"
        else:
            prompt = f"Fasse folgenden Text kurz zusammen:\n{frage}"
    
    elif user_input == "5":
        thema = input("Thema: ")
        plattform = input("Plattform (LinkedIn/Instagram): ")

        prompt = f"""
        Schreibe einen professionellen Social Media Post.

        Thema: {thema}
        Plattform: {plattform}

        Der Post soll:
        - Aufmerksamkeit erzeugen
        - klar strukturiert sein
        - Hashtags enthalten
        """
   
    elif user_input == "6":
        produkt = input("Produktname: ")
        zielgruppe = input("Zielgruppe: ")

        prompt = f"""
        Schreibe eine überzeugende und verkaufsstarke Produktbeschreibung.

        Produkt: {produkt}
        Zielgruppe: {zielgruppe}

        Die Beschreibung soll:
        - emotional ansprechend sein
        - Vorteile klar hervorheben
        - einen Call-to-Action enthalten
        - professionell klingen
        """
    
    else:
        print("Ungültige Eingabe")
        continue

#  KI ANFRAGE
    
    antwort = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[
            {"role": "user", "content": prompt}
        ]
    )

    print("\n--- Ergebnis ---\n")
    print(antwort.choices[0].message.content)
```
