```python
text = input("Gib einen Text ein: ")

print("\n--- Optionen ---")
print("1 = Zusammenfassung")
print("2 = Großbuchstaben")

choice = input("Was möchtest du tun? (1/2): ")

if choice == "1":
    result = text[:50]
elif choice == "2":
    result = text.upper()
else:
    result = "Ungültige Auswahl"

print("\n--- Ergebnis ---")
print(result)
```
